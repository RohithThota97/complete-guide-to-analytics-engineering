```python
import pandas as pd
df = pd.read_csv("/Users/rohiththota/Local-Repo/complete-guide-to-analytics-engineering/red_30_tech_us_sales.csv")
```

```python
print(df.columns)
```

```python
# Normalize column names first
df.columns = df.columns.str.replace("\u00a0", " ", regex=False).str.strip()

# Clean and convert Order Total to numeric
df["Order Total"] = (
    df["Order Total"]
    .astype(str)
    .str.replace(r'[\s"$,]', "", regex=True)
    .astype(float)
)
```

```python
df['OrderDate'] = pd.to_datetime(df['OrderDate'], format='%m/%d/%Y', errors='coerce')
```
