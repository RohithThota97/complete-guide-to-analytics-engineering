```python
import pandas as pd
df = pd.read_csv(
    "/Users/rohiththota/Local-Repo/complete-guide-to-analytics-engineering/red_30_tech_us_sales.csv"
)
```

```python
df.info()
df.head()
```

```python
sum_sales_quantity = df.groupby("EmpID ")["Quantity"].sum()
print(sum_sales_quantity)
```

```python
import matplotlib.pyplot as plt
sum_sales_quantity.plot(kind="bar",figsize=(10, 6),xlabel="EmpID",ylabel="Total Quantity",title="Sales Quantity by Employee",color="skyblue")
plt.show()
```

```python
mean_sales_quantity = df.groupby("EmpID ")["Quantity"].mean()
max_sales_quantity = df.groupby("EmpID ")["Quantity"].max()
min_sales_quantity = df.groupby("EmpID ")["Quantity"].min()
sorted_mean_sales_quantity = mean_sales_quantity.sort_values(ascending=False)

print(mean_sales_quantity)
print(max_sales_quantity)
print(min_sales_quantity)
print(sorted_mean_sales_quantity)

sorted_mean_sales_quantity.plot(
    kind="bar",
    figsize=(10, 6),
    xlabel="EmpID",
    ylabel="Total Quantity",
    title="Sales Quantity by Employee",
    color="skyblue",
)
plt.show()
```

```python
summary_metrics_df =pd.DataFrame({
    'EmpID ':sum_sales_quantity.index,
    'Sum Quantity':sum_sales_quantity.values,
    'Mean Quantity':mean_sales_quantity.values,
    'Max Quantity':max_sales_quantity.values,
    'Min Quantity':min_sales_quantity.values
})
print(summary_metrics_df)
```

```python
sales_associates_df = pd.read_csv(
    "/Users/rohiththota/Local-Repo/complete-guide-to-analytics-engineering/red_30_tech_us_sales_associates.csv"
)
```

```python
merged_df = merged_df.assign(
    EndDate=pd.to_datetime(merged_df["End date"].fillna("1/1/2020"))
)
```

```python
merged_df["Tenure"] = (
    pd.to_datetime(merged_df["EndDate"]) - pd.to_datetime(merged_df["Start Date"])
).dt.days

merged_df.info()

merged_df.head()
```

```python
merged_df=merged_df.assign(SalesPerDay=merged_df['Sum Quantity']/merged_df['Tenure'].replace(0,1))

merged_df.head(44)
```
