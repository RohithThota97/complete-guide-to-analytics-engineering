```python
import pandas as pd
df = pd.read_csv(
    "/Users/rohiththota/Local-Repo/complete-guide-to-analytics-engineering/red_30_tech_us_sales.csv")
```

```python
grouped_df= df.groupby('EmpID ')['Quantity'].sum().sort_values(ascending=False)
print(grouped_df)
```

```python
grouped_df = df.groupby("EmpID ")["Quantity"].sum().sort_values(ascending=True)
print(grouped_df)
```

```python
top_5_employees = grouped_df.nlargest(5)
print(top_5_employees)
```
